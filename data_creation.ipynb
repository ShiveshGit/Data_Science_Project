### Create headers

```python
with open('dataset/headers.txt', 'r') as f:
    lines = f.readlines()
description=[]
for line in lines:
    description.append(line.split('(')[1].split(')')[0].strip())
description.insert(24,'instance weight')
description[-1]="94 or 95"
description.append('below or above $50k')
description[37]="fill inc questionaire for veteran's admin"
description[19]="tax filer status"
description[7]="marital status"
description[6]="enrolled in edu last week"
description[2]="industry code"
description[3]="occupation code"
description=pd.DataFrame({'description':description})
description
```

```python
data.columns=description['description']
data.head()
data.to_csv("train.csv")
test.columns=description['description']
test.to_csv("test.csv")
```
